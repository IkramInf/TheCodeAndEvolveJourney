```python
import random
```

```python
random.seed(1)
data = [random.randint(200, 300) for _ in range(30)]
print(data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a boxplot
plt.figure(figsize=(8, 6))
G = sns.boxplot(data=data)
plt.title("Boxplot of Given Data")
plt.show()
```

```python
outlier = [10, 25, 100, 400, 450, 500]
new_data = data + outlier
print(new_data)
```

```python
# Create a boxplot
plt.figure(figsize=(8, 6))
G = sns.boxplot(data=new_data)
plt.title("Boxplot of Given New Data")
plt.show()
```

# Variance

Formula: \begin{align*}
\text{Variance } \sigma^2 & = \frac{1}{N} \sum_{i=1}^{N} \left(X_i - \mu\right)^2
\end{align*}


```python
# distance: square of mean to x
```

```python
N = len(data)
mean = sum(data) / N
print(mean)
```

```python
variance = sum([(x - mean) ** 2 for x in data]) / N
print(f"Variance: {variance}")
```

```python
import numpy as np
```

```python
var = np.var(data)
print(f"Variance using Numpy: {variance}")
```

# Standard Deviation

Formula: \begin{align*}
\text{std } \sigma & = \sqrt{\text{Variance}}
\end{align*}


```python
std = variance ** 0.5
print(f"STD: {std}")
```

```python
std = np.std(data)
print(f"STD using Numpy: {std}")
```
